### RNAi therapeutic implication layer 

```python
import pandas as pd
import os

data_processed = os.path.expanduser("~/Pan-Autoimmune-miRNA-ML/data/processed")
print("Libraries loaded")
```

```python
## Load shared pan-autoimmune miRNAs and anchor gene interactions
## This step identifies which shared miRNAs target PTPN22 and NLRP1
## — these are the most therapeutically relevant candidates

shared_mirnas = pd.read_csv(f"{data_processed}/shared_pan_autoimmune_miRNAs.csv")
network = pd.read_csv(f"{data_processed}/pan_autoimmune_network_interactions.csv")

anchor_genes = ['PTPN22', 'NLRP1']

## Find shared miRNAs that target anchor genes
rnai_candidates = network[
    (network['Target Gene'].isin(anchor_genes)) &
    (network['miRNA'].isin(shared_mirnas['miRNA']))
]

print(f"RNAi therapeutic candidates:")
print(rnai_candidates)
```

```python
## Save RNAi therapeutic candidates
rnai_candidates.to_csv(f"{data_processed}/RNAi_therapeutic_candidates.csv", index=False)

## Document the therapeutic implication
summary = {
    'anchor_gene': ['PTPN22', 'NLRP1'],
    'miRNA_candidate': ['hsa-miR-181a-5p', 'None identified in functional MTI'],
    'diseases_shared': ['Vitiligo, SLE, RA', 'Vitiligo, SLE, RA'],
    'therapeutic_rationale': [
        'Pan-autoimmune shared miRNA targeting upstream T-cell regulator',
        'No functional MTI found — highlights gap for future experimental validation'
    ]
}

summary_df = pd.DataFrame(summary)
summary_df.to_csv(f"{data_processed}/RNAi_therapeutic_summary.csv", index=False)
print(summary_df)
```
